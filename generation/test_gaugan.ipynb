```python
import numpy as np
from keras import utils
import matplotlib.pyplot as plt

from gaugan import GauganPredictor
```

```python
generator = GauganPredictor(
    'trained_models/generator.h5',
    'trained_models/encoder.h5'
)
```

```python
labels = np.zeros((1, 256, 256, 1))
labels = utils.to_categorical(labels, 25)
generated = generator(labels)[0]
plt.imshow(generated)
```
